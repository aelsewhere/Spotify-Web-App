Emotions gathered from [The Hoffman Institute](https://www.hoffmaninstitute.org/wp-content/uploads/Practices-FeelingsSensations.pdf). Copy & pasted this table above into a new file (emotion_table.md). This will make it easier to build the table.

```python
emotes = pd.read_csv('emotion_table.md', sep='|', skipinitialspace=True)
# delete row 0 (-------- row)
emotes = emotes.drop(0)
emotes = emotes.drop(columns=['Unnamed: 0', 'Unnamed: 5'])
emotes.columns = emotes.columns.str.strip()
emotes['Family 1'] = emotes['Family 1'].str.strip()
```

```python
emotes
```

Each family has an assigned color. As of right now I am only using Family 1 as an umbrella.
Here is the hex codes for each umbrella.

```python
emote_colors = {
    'Loving': '#E71881',
    'Accepting': '#D15DBF',
    'Tender': '#DD7594',
    'Joy': '#DA41A5',
    'Courageous': '#CE71C7',
    'Hopeful': '#E77776',
    'Curious': '#D359BB',
    'Grateful': '#D974A5',
    'Numb': '#EE795E',
    'Doubt': '#CC6FCF',
    'Stressed': '#E27685',
    'Guilt': '#C38477',
    'Powerless': '#D974A6',
    'Angry': '#ED7965',
    'Shame': '#6698AF',
    'Fragile': '#E1768D',
    'Fear': '#CE8170',
    'Sad': '#17A9E0'
}
```

```python
# print each emotion in its color

for emote in emote_colors:
    print(f'\033[38;2;{int(emote_colors[emote][1:3], 16)};{int(emote_colors[emote][3:5], 16)};{int(emote_colors[emote][5:], 16)}m{emote}\033[0m')
```

```python
print(emotes['Family 1'].unique())
```

```python
# add hex column to emote table
emotes['hex'] = emotes['Family 1'].map(emote_colors)
```

```python
# export emotes table to json -> locate in tables folder
emotes.to_json('emotes.json', orient='records')
```
